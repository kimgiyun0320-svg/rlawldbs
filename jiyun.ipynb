```python
import pandas as pd
import matplotlib.pyplot as plt

# cap 파일 불러오기
df = pd.read_csv("20250715Align_grid.cap", sep=',', header=None)
df.columns = ["Time", "X", "Y", "Intensity1", "Intensity2"]

# CSV로 저장 (선택 사항)
# df.to_csv("converted_output.csv", index=False)


```

```python
df.head()
```

```python
# X와 Y에서 문자 제거하고 숫자만 추출해서 float로 변환
df["X"] = df["X"].str.replace("X", "", regex=False).astype(float)
df["Y"] = df["Y"].str.replace("Y", "", regex=False).astype(float)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt

# Intensity1과 Intensity2를 각각 다른 색으로 산점도 그리기
plt.figure(figsize=(8, 6))
plt.scatter(df['X'], df['Y'], c=df['Intensity1'], cmap='Reds', label='Intensity1', alpha=0.6)
plt.scatter(df['X'], df['Y'], c=df['Intensity2'], cmap='Blues', label='Intensity2', alpha=0.6, marker='x')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Intensity1 & Intensity2 산점도')
plt.colorbar(label='Intensity')  # 색상바
plt.legend(['Intensity1', 'Intensity2'])
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))

# Intensity1 산점도
plt.subplot(1, 2, 1)
plt.scatter(df['X'], df['Y'], c=df['Intensity1'], cmap='Reds', alpha=0.7)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Intensity1 산점도')
plt.colorbar(label='Intensity1')

# Intensity2 산점도
plt.subplot(1, 2, 2)
plt.scatter(df['X'], df['Y'], c=df['Intensity2'], cmap='Blues', alpha=0.7)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Intensity2 산점도')
plt.colorbar(label='Intensity2')

plt.tight_layout()
plt.show()
```

```python
df1 = df[df['Intensity1'] >= 1]

plt.scatter(df1['X'], df1['Y'], c=df1['Intensity1'], cmap='Reds', alpha=0.7)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Intensity1 ≥ 1 산점도')
plt.colorbar(label='Intensity1')
plt.show()
```
