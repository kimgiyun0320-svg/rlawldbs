```python
# 입력 및 출력 파일 경로
file_path = "20250715Align_grid.cap"
output_path = file_path.replace("20250715Align_grid.cap", "20250715Align_grid_modified.cap")

# 수정된 파일 경로 (_modified 붙이기)
output_path = file_path.replace(".cap", "_modified.cap")

with open(file_path, 'r') as infile, open(output_path, 'w') as outfile:
    for line in infile:
        parts = line.strip().split(',')
        if len(parts) >= 4:
            try:
                fourth_value = float(parts[3])
                if fourth_value > 1:
                    outfile.write(line)  # 조건을 만족하면 출력
            except ValueError:
                continue  # 잘못된 숫자 형식이면 무시
```

```python
import matplotlib.pyplot as plt

# 파일 경로
file_path = "./20250715Align_grid.cap" # 긴 거

# 좌표 저장용 리스트
x_coords = []
y_coords = []

# 파일 파싱 (100개까지만)
with open(file_path, 'r') as file:
    for line in file:
        #if len(x_coords) >= 13128:
        #    break  # 100개까지만 처리

        parts = line.strip().split(',')
        if len(parts) >= 3:
            try:
                x = float(parts[1][1:])  # 'X1.992' -> 1.992
                y = float(parts[2][1:])  # 'Y2.479' -> 2.479
                x_coords.append(x)
                y_coords.append(y)
            except ValueError:
                continue

# 시각화
plt.figure(figsize=(8, 6))
plt.scatter(x_coords, y_coords, c='blue', marker='o', s=1)  # 점 크기 아주 작게
plt.title("Extracted X-Y Coordinates (first 100)")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.xlim(-100, 100)   # X축 범위 고정
plt.ylim(-80, 100)     # Y축 범위 고정
plt.show()
```

```python
import math

# 예시: 이미 수집된 x, y 좌표 리스트
# x_coords = [x1, x2, x3, ...]
# y_coords = [y1, y2, y3, ...]

def compute_average_distance(x_coords, y_coords):
    if len(x_coords) < 2:
        return 0.0  # 거리 계산 불가

    total_distance = 0.0
    count = 0

    for i in range(1, 13128):
        dy = y_coords[i] - y_coords[i - 1]
        total_distance += dy
        count += 1

    average = total_distance / count
    return average
```

```python
x_coords = []
y_coords = []

with open("20250715Align_grid_modified2.cap", 'r') as file:
    for line in file:
        parts = line.strip().split(',')
        if len(parts) >= 3:
            try:
                x = float(parts[1][1:])
                y = float(parts[2][1:])
                x_coords.append(x)
                y_coords.append(y)
            except ValueError:
                continue

avg_distance = compute_average_distance(x_coords, y_coords)
print(f"짧은 거 평균 거리: {avg_distance:.4f}")
```
